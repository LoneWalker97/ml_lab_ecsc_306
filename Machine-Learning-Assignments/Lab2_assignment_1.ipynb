```python

import tensorflow as tf
import numpy
import matplotlib.pyplot as plt

# Parameters
learning_rate = 0.005
training_epochs = 10
display_step = 10

# Training Data
train_X = numpy.asarray([1.,2.,3.,4.])
train_Y = numpy.asarray([5.,6.,7.,8.])
n_samples = train_X.shape[0]


W = 0
b = 0



# Construct a linear model



# Fit all training data
with tf.Session() as sess:
    for epoch in range(training_epochs):
        for (x, y) in zip(train_X, train_Y):
            pred = tf.add(tf.multiply(x, W), b)
            cost = tf.reduce_sum(tf.pow(pred-y, 2))/(2*n_samples)
            for num in train_X:
                temp_0 = b - learning_rate*sess.run(cost)
                temp_1 = W - learning_rate*sess.run(cost)*num
                b = temp_0
                W = temp_1
                print ("Theata_1 = ",W)
                print ("Theata_0 = ",b)
            
            

#Graphic display
    plt.plot(train_X, train_Y,'ro', label='Original data')
    plt.plot(train_X, W * train_X + b, label='Fitted line')
    plt.legend()
    plt.show()

print ("Theata_1 :",W,"\n" , "Theata_0 :",b)
```
